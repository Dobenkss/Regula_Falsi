```python
import numpy as np
import matplotlib.pyplot as plt

# Soal 1: Input fungsi dari pengguna
fungsi_str = input("Masukkan persamaan non-linear (contoh: x**3 - x - 2): ")
def f(x):
    return eval(fungsi_str)

# Soal 2-8: Fungsi Regula Falsi tetap sama
def regula_falsi(f, a, b, tol=1e-5, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Tidak ada akar di interval tersebut.")
        return None
    c = a
    for i in range(max_iter):
        print(f"iterasi: ke-{i+1}: c = {c}")
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        if abs(f(c)) < tol:
            break
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return c

# Soal 7: Meminta interval awal dari pengguna
a = float(input("Masukkan batas bawah interval (a): "))
b = float(input("Masukkan batas atas interval (b): "))

# Soal 8: Memanggil fungsi Regula Falsi dan mencari akar
akar = regula_falsi(f, a, b)
if akar is not None:
    print(f"Akar yang ditemukan: {akar}")

    # Soal 9: Plotting grafik
    x = np.linspace(a - 1, b + 1, 400)
    y = f(x)

    plt.plot(x, y, label=f'f(x) = {fungsi_str}')
    plt.axhline(0, color='black', lw=0.5, ls='--')
    plt.axvline(akar, color='red', lw=1, label=f'Akar: {akar:.5f}')

    plt.title('Grafik Fungsi dan Akar')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.show()
```
